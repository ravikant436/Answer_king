**Remove Stop Words**

```
def remove_dirty_chars(w):
  w = w.replace("'", "")
  return w
```

```
stop_words = set(stopwords.words('german'))
stop_words.update([",", ".", "'s", "@"])

filtered_wiki_tokens = [] 
filtered_legal_tokens = [] 

for w in wiki_tokens: 
    if w not in stop_words: 
        filtered_wiki_tokens.append(w) 

for w in legal_tokens: 
    if w not in stop_words: 
        filtered_legal_tokens.append(remove_dirty_chars(w)) 
```

```
import re
def hasNumbers(inputString):
  return bool(re.search(r'\d', inputString))
```

```
legal_tokens_dict = {}
unique_list = np.unique(np.array(filtered_legal_tokens))
for item in unique_list:
  if not hasNumbers(str(item)) and len(str(item)) > 2:
    item_count = filtered_legal_tokens.count(str(item))
    legal_tokens_dict[str(item)] = item_count
```

#Evaluate Average and Standard devaition of Weirdness and F-special

```
import pandas as pd
term_weirdness_data = pd.read_csv("/content/drive/My Drive/Colab Notebooks/AutomaticOnt/term_weirdness_data.csv")
del term_weirdness_data['Unnamed: 0']
term_weirdness_data.head()
```

```
#Calculate averages
n_special = len(term_weirdness_data)
list_term_frequency_slc = list(term_weirdness_data["Term_frequency_SLC"])
list_weirdness = list(term_weirdness_data["Weirdness"])
list_term = list(term_weirdness_data["Term"])

avg_f = sum(list_term_frequency_slc)/n_special
avg_weirdness = sum(list_weirdness)/n_special
print(avg_f)
print(avg_weirdness)
```

```
# calculate standard Devaiations
f_avg_diff_square_sum = 0
weirdness_avg_diff_square_sum = 0

for item in list_term_frequency_slc:
  f_avg_diff_square_sum += np.square(item - avg_f)

for item in list_weirdness:
  weirdness_avg_diff_square_sum += np.square(item - avg_weirdness)

sd_f = f_avg_diff_square_sum/(n_special*(n_special-1))
sd_weirdness = weirdness_avg_diff_square_sum/(n_special*(n_special-1))

print(sd_f)
print(sd_weirdness)
```

#Keyword Extraction

```
keywords = []
f_thresold = 0*sd_f
weirdness_thresold = 0*sd_weirdness

for ind in term_weirdness_data.index:
  z_freq_term = (term_weirdness_data['Term_frequency_SLC'][ind] - avg_f)/ sd_f
  z_weirdness_term = (term_weirdness_data['Weirdness'][ind] - avg_weirdness)/ sd_weirdness
  if z_freq_term > f_thresold and z_weirdness_term > weirdness_thresold:
    keywords.append(term_weirdness_data['Term'][ind])

keywords = np.unique(keywords)
len(keywords)
```

```
def checkKey(dict, key):      
    if key in dict.keys(): 
        return True
    else: 
        return False 
```

```
def get_collocation_list(word, corpus):
  collocation_dictionary = {}
  result_list = []
  for item in range(len(corpus)):
    total_sum = 0
    if corpus[item].lower() == word.lower():
      total_sum += 1
      i = -5
      inner_list = []
      while i<5:
        if not checkKey(collocation_dictionary, str(corpus[item+i])):
          #intialize
          collocation_dictionary[str(corpus[item+i])] = {}
          collocation_dictionary[str(corpus[item+i])][0] = 0
          collocation_dictionary[str(corpus[item+i])][1] = 0
          collocation_dictionary[str(corpus[item+i])][2] = 0
          collocation_dictionary[str(corpus[item+i])][3] = 0
          collocation_dictionary[str(corpus[item+i])][4] = 0
          collocation_dictionary[str(corpus[item+i])][5] = 0
          collocation_dictionary[str(corpus[item+i])][6] = 0
          collocation_dictionary[str(corpus[item+i])][7] = 0
          collocation_dictionary[str(corpus[item+i])][8] = 0
          collocation_dictionary[str(corpus[item+i])][9] = 0
          collocation_dictionary[str(corpus[item+i])][10] = 0
          collocation_dictionary[str(corpus[item+i])][11] = str(corpus[item+i])
          #update value
          collocation_dictionary[str(corpus[item+i])][i+5] += 1
          collocation_dictionary[str(corpus[item+i])][10] += 1
        else:
          # key already present
          collocation_dictionary[str(corpus[item+i])][i+5] += 1
          collocation_dictionary[str(corpus[item+i])][10] += 1
        #inner_list.append(corpus[item+i])
        i += 1
      #result_list.append(inner_list)
  return collocation_dictionary
```
