**Named Entity Recognition**

```
for ent in doc.ents:
  print(ent.text, ent.label_) 
```

#Text Lemmatization using Tiger Corpus

Source: https://github.com/WZBSocialScienceCenter/germalemma

N...' (nouns)
'V...' (verbs)
'ADJ...' (adjectives)
'ADV...' (adverbs)

```
pip install -U germalemma
```

```
from germalemma import GermaLemma
lemmatizer = GermaLemma()
```

```
# passing the word and the POS tag ("N" for noun, "ADJ" for Adjective, "V" for verb, "ADV" for adverb)
lemma = lemmatizer.find_lemma('Steuerpflichtige', 'ADJ')
print(lemma)
```
