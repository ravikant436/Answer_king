```
nlp = de_core_news_md.load()
```

```
def getPos(word):
  text = (word) 
  doc = nlp(text)
  for token in doc: 
    if token.pos_ == "NOUN" or token.pos_ == "PROPN" or token.pos_ == "X":
      return 'N'
    elif token.pos_ == "VERB":
      return 'V'
    elif token.pos_ == "ADJ" or token.pos_ == "ADV":
      return token.pos_
    else:
      return 'N'
```

```
def getLemma(word):
  word_pos = getPos("" + word + "")
  print(word_pos)
  lemma = lemmatizer.find_lemma(word, word_pos)
  return lemma
```

#Get all nodes and all the edge sets

```
def getHypernymEdges(df):
  edge_set = []
  for index, row in df.iterrows():
    if row["Hypernyms"] != "":
      temp_str = row["Hypernyms"][1:-1]
      temp_arr = temp_str.split(',')
      for item in temp_arr:
        item = item.strip()
        item = item[1:-1]
        temp_list = [item, row["Lemma_word"]]
        edge_set.append(temp_list)
  return edge_set
```

```
def getHyponymEdges(df):
  edge_set = []
  for index, row in df.iterrows():
    if row["Hyponyms"] != "":
      temp_str = row["Hyponyms"][1:-1]
      temp_arr = temp_str.split(',')
      for item in temp_arr:
        item = item.strip()
        item = item[1:-1]
        temp_list = [row["Lemma_word"], item]
        edge_set.append(temp_list)
  return edge_set
```

```
hyper_edge_set_df = hyper_df[["Lemma_word", "Hypernyms"]].replace(np.nan, '')
hyper_edge_set = getHypernymEdges(hyper_edge_set_df)
```

```
hypo_edge_set_df = hypo_df[["Lemma_word", "Hyponyms"]].replace(np.nan, '')
hypo_edge_set = getHyponymEdges(hypo_edge_set_df)
```

```
temp = hyper_edge_set + hypo_edge_set
all_edges_df = pd.DataFrame(temp, columns=["start", "End"])
all_edges_df = all_edges_df.drop_duplicates()
all_edges_df.shape
```

**Get Node set**

```
all_nodes = []
for index, row in all_edges_df.iterrows():
  all_nodes.append(row["start"])
  all_nodes.append(row["End"])
all_nodes_df = pd.DataFrame(all_nodes, columns=["Nodes"])
all_nodes_df = all_nodes_df.drop_duplicates()
all_nodes_df.shape
```
